## Running the analysis

Here you can see how the code is imported from the `.py` files and implemented to conduct the analysis.

```python
# Import modules to estimate and show results

import bases 
import searchea as sc
import matplotlib.pyplot as plt

# Reasing the data
kills = bases.extract_kills('../assignment-final-data/kills.txt')
cheaters = bases.extract_cheaters('../assignment-final-data/cheaters.txt')

# Finding motifs in the original network
observers = sc.find_observers(kills, cheaters)
original_motifs = sc.find_motifs(observers, cheaters)

# Conducting 10 shufflings and adding the number of motifs to one list
shuffled_list = []
for i in range(10):
    sc.shuffle_network(kills, cheaters)
    stuff_observers = sc.find_observers(kills, cheaters)
    shuff_motifs = sc.find_motifs(stuff_observers, cheaters)
    shuffled_list.append(len(shuff_motifs))

# Plotting the results
plt.figure(figsize=[10,8])

plt.axvline(x = len(original_motifs), c = 'darkred')
plt.hist(shuffled_list)
plt.text(len(original_motifs), 1, "N of original motifs", rotation = 90, backgroundcolor='darkred', color = 'white')
plt.xlabel('Number of motifs')
plt.ylabel('Frequency')
plt.title('Distribution of motif numbers in original and randomly shuffled networks')
plt.show()
```

**Basing on the analysis conducted and the graph above, in particular, we can see, that the number of motifs in the randomised networks is only slightly different then the number in the original network. Considering that the change in the number of motifs is in range of 15 motifs, we can conclude that observing a cheater _or_ being killed by one does not "infect" a person with cheating. Comparing the original universe to the randomised ones we can conclude that the "contagions" happened in the original network can be seen as arbitrary.**
